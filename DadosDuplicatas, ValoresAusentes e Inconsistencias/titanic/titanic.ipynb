```python
print("Limpando valores nulos das colunas Age, Cabin e Embarked")
df = df.dropna(subset=['Age'], axis=0)
df = df.dropna(subset=['Cabin'], axis=0)
df = df.dropna(subset=['Embarked'], axis=0)
```

```python
print(df.info())
print("padronizando as colunas de Sex e embarked:")
#Textos
#mudando para valor STR
df['Embarked'] = df['Embarked'].astype(str)
df['Sex'] = df['Sex'].astype(str)
#Removendo espaços em brancos no inicio e fim
df['Embarked_Strip'] = df['Embarked'].str.strip()
df['Sex_Strip'] = df['Sex'].str.strip()

#primeira letra maiscula
df['Embarked_Title'] = df['Embarked'].str.strip(). str.title()
df['Sex_Title'] = df['Sex'].str.strip(). str.title()
```
